```python
import numpy as np
import matplotlib.pyplot as plt

pi = np.linspace(0,1,100)
var = pi*(1-pi)
plt.plot(pi,var)
plt.title('Funcion de varianza de distribciones Bernoulli')
plt.xlabel( '$\pi$')
plt.ylabel('Var')
plt.grid()
plt.show()
```

```python
fig=plt.figure(figsize=(10, 8))

plt.title('Informacin de Shannon en la distribucin de Bernoulli de acuerdo al parmetro $p$',fontsize=14)
p=np.linspace(1e-3,1-1e-3,100)
I_1=-np.log2(p)
I_0=-np.log2(1-p)
plt.plot(p,I_1,label=r"$I(1)$")
plt.plot(p,I_0,label=r"$I(0)$")
plt.xlabel("$p$", fontsize = 14)
plt.ylabel("$I(p) = -\log \ p$", fontsize = 14)
plt.axvline(0.5, 0, 10,marker='d',color='k')
plt.legend()
plt.grid()
```

```python
fig=plt.figure(figsize=(10,8))
H_ber=lambda X: -np.sum(X*np.log2(X),axis=1) # Function to estimate the entropy of a bernoulli distribution
p=np.linspace(1e-3,1-1e-3,100)
probs_X=np.vstack([p,1-p]).T
plt.plot(p,H_ber(probs_X),'r')
plt.xlabel("$p$",fontsize =14)
plt.ylabel("$H(X)$",fontsize =14)
plt.ylim([0,1.05])
plt.xlim([0,1])
plt.title('Entropa de las distirbuciones de la familia Bernoulli',fontsize =14)
plt.axvline(0.5, 0, 10,marker='d',color='k')
plt.grid()
```
